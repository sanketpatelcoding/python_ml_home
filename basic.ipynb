```python
# 1. Import the libraries we need
import numpy as np                  # helps with numbers/arrays
import matplotlib.pyplot as plt     # draws pictures
from sklearn.linear_model import LinearRegression   # the magic tool
# 2. Our tiny example data
# X must be 2D array → that's why we write .reshape(-1, 1)
X = np.array([1, 2, 3, 5, 6, 8]).reshape(-1, 1)     # hours
y = np.array([30, 45, 55, 75, 85, 95])              # scores
# 3. Create the model (like an empty box)
model = LinearRegression()

# 4. Teach (train) the model with the data
model.fit(X, y)

# 5. Now the model knows the best line!
# Let's see what it learned:
print("Intercept (a)  :", model.intercept_)
print("Slope (b)      :", model.coef_[0])

# Example prediction: what score if someone studies 4 hours?
hours_new = np.array([[4]])          # again 2D shape
predicted = model.predict(hours_new)
print(f"If study 4 hours → predicted score ≈ {predicted[0]:.0f}")

# Draw the original points
plt.scatter(X, y, color='blue', label='Real data')

# Draw the line the model found
X_line = np.array([0, 10]).reshape(-1, 1)      # from 0 to 10 hours
y_line = model.predict(X_line)
plt.plot(X_line, y_line, color='red', label='Best fit line')

plt.xlabel('Hours studied')
plt.ylabel('Exam score')
plt.title('Linear Regression Example')
plt.legend()
plt.grid(True)
plt.show()
```
