```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score, classification_report

# 
# 1. Load data
# 
url = "https://raw.githubusercontent.com/npradaschnor/Pima-Indians-Diabetes-Dataset/master/diabetes.csv"
data = pd.read_csv(url)

print("First 3 rows:")
print(data.head(3))
print("\nShape:", data.shape)
print("\nColumns:", data.columns.tolist())

# 
# 2. Handle invalid zeros (very important for this dataset!)
# 
cols_with_invalid_zeros = ['Glucose', 'BloodPressure', 'SkinThickness', 'Insulin', 'BMI']

print("\nMedian values before imputation:")
print(data[cols_with_invalid_zeros].median())

# Replace 0  median (only in selected columns)
for col in cols_with_invalid_zeros:
    data[col] = data[col].replace(0, data[col].median())

print("\nAfter imputation - no more zeros in critical columns:")
print((data[cols_with_invalid_zeros] == 0).sum())

# 
# 3. Features & target
# 
X = data.drop('Outcome', axis=1)
y = data['Outcome']

# 
# 4. Train / test split
# 
X_train, X_test, y_train, y_test = train_test_split(
    X, y, 
    test_size=0.25, 
    random_state=42, 
    stratify=y          # important for imbalanced classes
)

# 
# 5. Scaling (SVMs are very sensitive to feature scales)
# 
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled  = scaler.transform(X_test)

# 
# 6. Visualization - Glucose vs BMI (before & after scaling)
# 
fig, axes = plt.subplots(1, 2, figsize=(14, 6))

# Before scaling
axes[0].scatter(
    X_train['Glucose'], 
    X_train['BMI'], 
    c=y_train, 
    cmap='coolwarm', 
    alpha=0.6, 
    edgecolor='k', 
    s=60
)
axes[0].set_title("Before Scaling\nGlucose vs BMI (train set)")
axes[0].set_xlabel("Glucose")
axes[0].set_ylabel("BMI")

# After scaling
axes[1].scatter(
    X_train_scaled[:, 1],          # Glucose is column index 1
    X_train_scaled[:, 5],          # BMI is column index 5
    c=y_train, 
    cmap='coolwarm', 
    alpha=0.6, 
    edgecolor='k', 
    s=60
)
axes[1].set_title("After Standard Scaling")
axes[1].set_xlabel("Scaled Glucose")
axes[1].set_ylabel("Scaled BMI")

plt.tight_layout()
plt.show()

# 
# 7. Train SVC model
# 
model = SVC(
    kernel='rbf',      # try also: 'linear', 'poly'
    C=1.0,             # regularization (higher = less regularization)
    gamma='scale',     # kernel coefficient
    random_state=42,
    class_weight='balanced'   # helps with class imbalance
)

model.fit(X_train_scaled, y_train)

# 
# 8. Evaluate
# 
y_pred = model.predict(X_test_scaled)

acc = accuracy_score(y_test, y_pred)
print(f"\nTest Accuracy: {acc:.4f}  ({acc*100:.2f}%)")
print("\nClassification Report:")
print(classification_report(y_test, y_pred, target_names=['No Diabetes (0)', 'Diabetes (1)']))

# Bonus: how many support vectors?
print(f"\nNumber of support vectors: {model.n_support_}")
print(f"Support vectors per class: {model.n_support_}")
```
